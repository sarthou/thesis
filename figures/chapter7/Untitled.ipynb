```python
#x_max = max(bb["ucs_time(ms)"].describe().max(), bb["ids_time(ms)"].describe().max(), bb["dfs_time(ms)"].describe().max()) + 10
ucs_max = bb["original_time(ms)"].describe().max()

x_max = 100
xs = np.linspace(0, x_max, 8000)
ys_ucs = [len(bb_augmented[bb_augmented["original_time(ms)"] < x].index) * 100 / NB_OBJECTS for x in xs]
ys_ids = [len(bb_augmented[bb_augmented["tasks_time(ms)"] < x].index) * 100 / NB_OBJECTS for x in xs]
ys_dfs = [len(bb_augmented[bb_augmented["CR_time(ms)"] < x].index) * 100 / NB_OBJECTS for x in xs]
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(10, 6, forward=True)
plt.plot(xs, ys_ucs, '#1f77b4', linestyle="--")
plt.plot(xs, ys_ids, '#2ca02c', linestyle="-.")
plt.plot(xs, ys_dfs, '#ff7f0e', linestyle="-")
plt.grid(axis='x', which="minor", linestyle="--", linewidth=0.5)
plt.grid(axis='both', which="major", linestyle="--", linewidth=1)
vals = ax.get_yticks()
print(vals)
ax.set_yticklabels(['{:0.0f}%'.format(x) for x in vals[1:]])
#plt.plot([ucs_max, ucs_max], [-10, 105], '--', lw=1)
plt.xscale("log")
plt.legend(("Original", "Past tasks", "Compound realtions"))
plt.xlabel("Time (ms)")
plt.ylabel("Percentage of individuals disambiguated")
plt.ylim(0, 105)
plt.show()
```
